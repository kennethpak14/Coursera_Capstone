## 1. Initial scraping and parsing of data
Using the [Requests](https://requests.readthedocs.io/en/master/), [BeautifulSoup](https://beautiful-soup-4.readthedocs.io/en/latest/), and [Pandas](https://pandas.pydata.org/docs/) packages, we scrape the postal code data and assign it to a dataframe.

```python
import requests # to grab the html from the website
from bs4 import BeautifulSoup # to do initial cleaning and parsing of data
import pandas as pd # to store data
import numpy as np
```

First, we grab the HTML from the website using requests and store it in a BeautifulSoup object

```python
# Grabbing the data and reading it through BeautifulSoup
url = "https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M"
wikipage = requests.get(url)

soup = BeautifulSoup(wikipage.content, 'lxml')
```

We can view the HTML structure more clearly with *'.prettify()'*

```python
print(soup.prettify())
```

Once we understand the key tags that make up the table we want to scrape (e.g. *'tbody'*, *'th'*, *'tr'*, etc.), we can go through the table and store the data into two lists: the **data** and the **headers**.

```python
data = []
headers = []
table = soup.find('table', class_ = 'wikitable sortable')
table_body = table.find('tbody')

# Adding headers of table into a list
table_headers = table.find_all('th')
for h in table_headers:
    headers.append(h.text.strip())

# Adding the body of table to another list
rows = table_body.find_all('tr')
for row in rows:
    cols = row.find_all('td')
    data.append([ele.text.strip() for ele in cols if ele])
```

We store the data in a dataframe and clean up the data a bit more.

```python
df = pd.DataFrame(data, columns = headers)

df = df[1:] # removing header row from df
df = df[df.Borough != "Not assigned"].reset_index(drop = True) # removing 'Not assigned' boroughs
```

Now there are no more boroughs or neighborhoods that are "Not assigned" in our df.

First, test running with the first neighborhood area, Regent Park/Harbourfront.

Getting (up to) the top 100 venues nearby Regent Park & Harbourfront

```python
LIMIT = 2 # num of venues grabbed
radius = 500 # meters nearby

url = 'https://api.foursquare.com/v2/venues/explore?&client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}'.format(
    CLIENT_ID,
    CLIENT_SECRET,
    VERSION,
    neigh_lat,
    neigh_long,
    radius,
    LIMIT)

results = requests.get(url).json()
results # viewing the json file
```

```python
# function that extracts the category of the venue from json file
def get_category_type(row):
    try:
        categories_list = row['categories']
    except:
        categories_list = row['venue.categories']
        
    if len(categories_list) == 0:
        return None
    else:
        return categories_list[0]['name']
```

```python
venues = results['response']['groups'][0]['items']
    
nearby_venues = json_normalize(venues) # flatten JSON

# filter columns
filtered_columns = ['venue.name', 'venue.categories', 'venue.location.lat', 'venue.location.lng']
nearby_venues = nearby_venues.loc[:, filtered_columns]

# filter the category for each row
nearby_venues['venue.categories'] = nearby_venues.apply(get_category_type, axis=1)

# clean columns
nearby_venues.columns = [col.split(".")[-1] for col in nearby_venues.columns]

nearby_venues.head()
```

Defining a function to automate the venue API call and returning the results in a dataframe.

```python
def getNearbyVenues(names, latitudes, longitudes, radius=500):
    
    venues_list=[]
    for name, lat, lng in zip(names, latitudes, longitudes):
        print(name)
            
        # create the API request URL
        url = 'https://api.foursquare.com/v2/venues/explore?&client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}'.format(
            CLIENT_ID, 
            CLIENT_SECRET, 
            VERSION, 
            lat, 
            lng, 
            radius, 
            LIMIT)
            
        # make the GET request
        results = requests.get(url).json()['response']['groups'][0]['items']
        
        # return only relevant information for each nearby venue
        venues_list.append([(
            name, 
            lat, 
            lng, 
            v['venue']['name'], 
            v['venue']['location']['lat'], 
            v['venue']['location']['lng'],  
            v['venue']['categories'][0]['name']) for v in results])

    nearby_venues = pd.DataFrame([item for venue_list in venues_list for item in venue_list])
    nearby_venues.columns = ['Neighborhood', 
                  'Neighborhood Latitude', 
                  'Neighborhood Longitude', 
                  'Venue', 
                  'Venue Latitude', 
                  'Venue Longitude', 
                  'Venue Category']
    
    return(nearby_venues)
```

```python
toronto_venues = getNearbyVenues(names=t_df['Neighborhood'],
                                   latitudes=t_df['Latitude'],
                                   longitudes=t_df['Longitude'])
```

```python
# Viewing the results of the API call
print(toronto_venues.shape)
toronto_venues.head()
```

### 3c. Segmenting the neighborhoods based on venue categories 

```python
# Viewing count of venues per neighborhood
toronto_venues.groupby('Neighborhood').count()
```

```python
# one hot encoding
toronto_onehot = pd.get_dummies(toronto_venues[['Venue Category']], prefix="", prefix_sep="")

# add neighborhood column back to dataframe
toronto_onehot['Neighborhood'] = toronto_venues['Neighborhood'] 

# move neighborhood column to the first column
fixed_columns = [toronto_onehot.columns[-1]] + list(toronto_onehot.columns[:-1])
toronto_onehot = toronto_onehot[fixed_columns]

toronto_onehot.head()
```

We can find the density or frequency of venue type by grouping the data and finding the mean values.

```python
toronto_grouped = toronto_onehot.groupby('Neighborhood').mean().reset_index()
toronto_grouped
```

By looking at the magnitude of the values, we can determine the most frequent types of venues in a neighborhood.

```python
num_top_venues = 5

for hood in toronto_grouped['Neighborhood']:
    print("----"+hood+"----")
    temp = toronto_grouped[toronto_grouped['Neighborhood'] == hood].T.reset_index()
    temp.columns = ['venue','freq']
    temp = temp.iloc[1:]
    temp['freq'] = temp['freq'].astype(float)
    temp = temp.round({'freq': 2})
    print(temp.sort_values('freq', ascending=False).reset_index(drop=True).head(num_top_venues))
    print('\n')
```

```python
def return_most_common_venues(row, num_top_venues):
    row_categories = row.iloc[1:]
    row_categories_sorted = row_categories.sort_values(ascending=False)
    
    return row_categories_sorted.index.values[0:num_top_venues]
```

```python
num_top_venues = 10

indicators = ['st', 'nd', 'rd']

# create columns according to number of top venues
columns = ['Neighborhood']
for ind in np.arange(num_top_venues):
    try:
        columns.append('{}{} Most Common Venue'.format(ind+1, indicators[ind]))
    except:
        columns.append('{}th Most Common Venue'.format(ind+1))

# create a new dataframe
neighborhoods_venues_sorted = pd.DataFrame(columns=columns)
neighborhoods_venues_sorted['Neighborhood'] = toronto_grouped['Neighborhood']

for ind in np.arange(toronto_grouped.shape[0]):
    neighborhoods_venues_sorted.iloc[ind, 1:] = return_most_common_venues(toronto_grouped.iloc[ind, :], num_top_venues)

neighborhoods_venues_sorted.head()
```

### 3d. Clustering the neighborhoods by venue category

```python
from sklearn.cluster import KMeans

# set number of clusters
kclusters = 5

toronto_grouped_clustering = toronto_grouped.drop('Neighborhood', 1)

# run k-means clustering
kmeans = KMeans(n_clusters=kclusters, random_state=0).fit(toronto_grouped_clustering)

# check cluster labels generated for each row in the dataframe
kmeans.labels_[0:10] 
```

```python
# add clustering labels
neighborhoods_venues_sorted.insert(0, 'Cluster Labels', kmeans.labels_)

toronto_merged = t_df

# merge toronto_grouped with toronto_data to add latitude/longitude for each neighborhood
toronto_merged = toronto_merged.join(neighborhoods_venues_sorted.set_index('Neighborhood'), on='Neighborhood')

toronto_merged # check the last columns!
```

### 3e. Viewing the results

```python
# create map
map_clusters = folium.Map(location=[latitude, longitude], zoom_start=11)

# set color scheme for the clusters
x = np.arange(kclusters)
ys = [i + x + (i*x)**2 for i in range(kclusters)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# add markers to the map
markers_colors = []
for lat, lon, poi, cluster in zip(toronto_merged['Latitude'], toronto_merged['Longitude'], toronto_merged['Neighborhood'], toronto_merged['Cluster Labels']):
    label = folium.Popup(str(poi) + ' Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(map_clusters)
       
map_clusters
```

## 4. Interpretation of Results
The most obvious point to note about the analysis is that the clustering sizes were very skewed. The first cluster contains most of the data points, and this can be potentially attributed to the separation of urban and suburban areas. Taking a look at the most common venues in the first cluster, we see that those are values primarily attributed to more urban areas of living: cafes, restaurants, gift shops, hotels. In contrast, the most common venues in the other neighborhoods seem like suburban venues: gardens, parks, swim school.

If I were to do the analysis again, I would either add more boroughs or only focus on one. Adding more boroughs would give a more diverse range to the data (i.e. not just urban neighborhoods), and would allow the clusters to be of more sigificance. Currently, the second to fifth clusters act more as outlier values to the first cluster. Focusing on one borough, most likely Downtown Toronto, would allow the clusters to better differentiate between the nuances of that one borough.

## Introduction 

When a company or individual is looking to open a new store, location plays a strong role in the early success of a budding enterprise. While outstanding products and quality service are an essential part in retaining customers, with an unfavorable location, the venue might not gain enough traction for the business to start. The location of a venue determines a large portion of the visibility, foot traffic, and potential competitors.

That being said, this problem is a broad issue. There are a lot of nuances in determining the optimal spot, and it would not be as simple as running a quick algorithm; however, there are ways that the research could be narrowed down, and this analysis is one method in doing so.

For the scope of this analysis, the different neighborhoods of San Francisco will be segmented and clustered based on restaurant data to recommend potential desirable areas for opening a new restaurant. 

---
## Data

In this analysis, there will be *three* types of data:

**Neighborhood data**: Data will be scraped from [this website](http://zipatlas.com/us/ca/san-francisco/zip-code-comparison/median-household-income.htm) to get a list of postal codes, and their respective neighborhoods, and population data.

**Geolocation data**: Using the geocoder package, the geolocation of all the neighborhoods are grabbed to give proper coordinates when using the Foursquared API to search for venues.

**Venue data**: Venues nearby to the neighborhood will be grabbed using the Foursquared API. The nearby venues will be coupled with the neighborhood locations to cluster the neighborhoods in distinct groups.

```python
import requests # to grab the html from the website
from bs4 import BeautifulSoup # to do initial cleaning and parsing of data
import pandas as pd # to store data
import numpy as np
from geopy.geocoders import Nominatim
```

Extracting data on the different zip codes in SF using *BS4*:

```python
url = "http://zipatlas.com/us/ca/san-francisco/zip-code-comparison/median-household-income.htm"
data = requests.get(url)

soup = BeautifulSoup(data.content, "lxml")
print(soup.prettify())
```

```python
# Parsing data into lists
headers = []
data = []
row_data = []

table_headers = soup.find_all('td', class_ = 'report_header')
table_body = soup.find_all('td', class_ = 'report_data')

for h in table_headers:
    headers.append(h.text.strip())

num_col = 7
for row in table_body:
    if (num_col > 0):
        row_data.append(row.text.strip())
        num_col -= 1
    else:
        data.append(row_data)
        row_data = []
        num_col = 7 - 1
        row_data.append(row.text.strip())
```

Next, the obtained data is put into a dataframe and cleaned up some more:

```python
# Putting list into df
df = pd.DataFrame(data, columns = headers)

# Cleaning up data
df[['Latitude', 'Longitude']] = df.Location.str.split(", ", expand = True)
df.rename(columns = {'Zip Code': 'Zipcode', 'Avg. Income/H/hold': 'Avg HH Income'},inplace = True)
df.drop(columns = ['#','City','National Rank','Location'],inplace = True)
df['Population'].replace(',','',regex = True, inplace = True)
df['Avg HH Income'].replace(',','',regex = True, inplace = True)
df['Avg HH Income'].replace('\$','',regex = True,inplace = True)
df['Avg HH Income'].replace('.00','', regex = True,inplace = True)

# changing types
df = df.apply(pd.to_numeric)
df["Zipcode"]=df["Zipcode"].astype("str") #needs to be str to join
```

```python
display(df.head())
print(df.dtypes)
```

Grabbing SF neighborhood names with respective zipcodes from [this kaggle dataset](https://www.kaggle.com/savargaonkar/sf-zipcodes-limited/data) to merge with the pre-existing dataframe:

```python
# Reading kaggle csv data
file_path = 'your file path to downloaded Kaggle dataset'

# cleaning data
neigh_data = pd.read_csv(file_path)
neigh_data.drop(columns = ['Unnamed: 2', 'Unnamed: 3'], inplace = True)
neigh_data['Neighborhood'] = neigh_data['Neighborhood'].str.strip()
neigh_data.drop_duplicates(inplace = True)

# changing types
neigh_data["Zipcode"] = neigh_data["Zipcode"].astype("str")
```

```python
# Merging the neighborhoods w/ same zipcodes into same rows
clean_neigh_data = neigh_data.groupby('Zipcode').agg({'Neighborhood': lambda x: ', '.join(x)})

# Joining both tables to get all the needed data in one table
sf_df = df.merge(clean_neigh_data, left_on = "Zipcode", right_on = "Zipcode")

# Rearranging columns
sf_df = sf_df[['Zipcode', 'Neighborhood','Latitude','Longitude','Population','Avg HH Income']]
```

```python
display(sf_df)
```

Now that the base dataset is formed, the Foursquare data is obtained using the Foursquare API to get nearby venues for each neighborhood:

```python
CLIENT_ID = ''
CLIENT_SECRET = ''
VERSION = '20200705'
```

```python
# function to take the json file and gets the categories
def get_category_type(row):
    try:
        categories_list = row['categories']
    except:
        categories_list = row['venue.categories']
        
    if len(categories_list) == 0:
        return None
    else:
        return categories_list[0]['name']
```

```python
# function for making the API call and returning a dataframe with **ONLY FOOD VENUES**
# https://developer.foursquare.com/docs/api-reference/venues/search/
def getNearbyVenues(names, latitudes, longitudes, LIMIT, radius=400):
    
    venues_list=[]
    for name, lat, lng in zip(names, latitudes, longitudes):
        print(name)
            
        # create the API request URL
        url = 'https://api.foursquare.com/v2/venues/explore?&client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}&categoryId={}'.format(
            CLIENT_ID, 
            CLIENT_SECRET, 
            VERSION, 
            lat, 
            lng, 
            radius, 
            LIMIT,
            '4d4b7105d754a06374d81259') # this is the category ID for all food venues
            
        # make the GET request
        results = requests.get(url).json()['response']['groups'][0]['items']
        
        # return only relevant information for each nearby venue
        venues_list.append([(
            name, 
            lat, 
            lng, 
            v['venue']['name'], 
            v['venue']['location']['lat'], 
            v['venue']['location']['lng'],  
            v['venue']['categories'][0]['name']) for v in results])

    nearby_venues = pd.DataFrame([item for venue_list in venues_list for item in venue_list])
    nearby_venues.columns = ['Neighborhood', 
                  'Neighborhood Latitude', 
                  'Neighborhood Longitude', 
                  'Venue', 
                  'Venue Latitude', 
                  'Venue Longitude', 
                  'Venue Category']
    
    return(nearby_venues)
```

```python
sf_venues = getNearbyVenues(names=sf_df['Neighborhood'],
                                   latitudes=sf_df['Latitude'],
                                   longitudes=sf_df['Longitude'],
                                   LIMIT = 50)
```

```python
sf_venues.dtypes
```

```python
# Seeing count of each neighborhood
sf_venues_count = sf_venues.groupby('Neighborhood').count()
sf_venues_count
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
#sf_venues_count.hist(by = ['Neighborhood'])

fig_dims = (13, 8)
fig, ax = plt.subplots(figsize=fig_dims)
ax = sns.countplot(y='Neighborhood', 
                   ax=ax, 
                   data=sf_venues,
                   order = sf_venues['Neighborhood'].value_counts().index)

```

```python
# finding all neighborhoods with less than or equal to 10 to remove outliers from Foursquared venue data
removing_rows = list(sf_venues_count.index[sf_venues_count['Neighborhood Latitude'] <= 10])
removing_rows.append('St. Francis Wood, Miraloma, West Portal')
removing_rows.append('Visitacion Valley, Sunnydale')
print(removing_rows)

for row in removing_rows:
    for value, i in zip(sf_venues.Neighborhood, sf_venues.index):
        if (value == row):
            sf_venues.drop(i, inplace = True)
    
display(sf_venues)
```

```python
# removing those neighborhoods from the base dataset as well
for (index, colData) in sf_df.Neighborhood.iteritems():
    if (colData in removing_rows):
        sf_df.drop(index, inplace = True)

display(sf_df)
```

Now, the Foursquared venue data can be quantified using one hot encoding to determine venue frequency in a neighborhood:

```python
# one hot encoding
sf_onehot = pd.get_dummies(sf_venues[['Venue Category']], prefix="", prefix_sep="")

# add neighborhood column back to dataframe
sf_onehot['Neighborhood'] = sf_venues['Neighborhood'] 

# move neighborhood column to the first column
fixed_columns = [sf_onehot.columns[-1]] + list(sf_onehot.columns[:-1])
sf_onehot = sf_onehot[fixed_columns]

display(sf_onehot)
```

```python
sf_grouped = sf_onehot.groupby('Neighborhood').mean().reset_index()
sf_grouped.head()
```

The data can be viewed tabularly as such:

```python
num_top_venues = 5

for hood in sf_grouped['Neighborhood']:
    print("----"+hood+"----")
    temp = sf_grouped[sf_grouped['Neighborhood'] == hood].T.reset_index()
    temp.columns = ['venue','freq']
    temp = temp.iloc[1:]
    temp['freq'] = temp['freq'].astype(float)
    temp = temp.round({'freq': 2})
    print(temp.sort_values('freq', ascending=False).reset_index(drop=True).head(num_top_venues))
    print('\n')
```

```python
def return_most_common_venues(row, num_top_venues):
    row_categories = row.iloc[1:]
    row_categories_sorted = row_categories.sort_values(ascending=False)
    
    return row_categories_sorted.index.values[0:num_top_venues]
```

```python
num_top_venues = 10

indicators = ['st', 'nd', 'rd']

# create columns according to number of top venues
columns = ['Neighborhood']
for ind in np.arange(num_top_venues):
    try:
        columns.append('{}{} Most Common Venue'.format(ind+1, indicators[ind]))
    except:
        columns.append('{}th Most Common Venue'.format(ind+1))

# create a new dataframe
neighborhoods_venues_sorted = pd.DataFrame(columns=columns)
neighborhoods_venues_sorted['Neighborhood'] = sf_grouped['Neighborhood']

for ind in np.arange(sf_grouped.shape[0]):
    neighborhoods_venues_sorted.iloc[ind, 1:] = return_most_common_venues(sf_grouped.iloc[ind, :], num_top_venues)

display(neighborhoods_venues_sorted.head())
```

Now that the neighborhoods have been segmented, the dataset needs to be clustered; but first, the Silhoutte Score is used to determine the best number of clusters to use:

```python
from sklearn.metrics import silhouette_samples, silhouette_score
from sklearn.cluster import KMeans

max_range = 10
indices = []
scores = []

sf_grouped_c = sf_grouped.drop('Neighborhood', 1)

for kclusters in range(2, max_range) :
    
    # Run k-means clustering
    kmeans = KMeans(n_clusters = kclusters, init = 'k-means++', random_state = 0).fit_predict(sf_grouped_c)
    
    # Gets the score for the clustering operation performed
    score = silhouette_score(sf_grouped_c, kmeans)
    
    # Appending the index and score to the respective lists
    indices.append(kclusters)
    scores.append(score)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

def plot(x, y, xlabel, ylabel):
    plt.figure(figsize=(20,10))
    plt.plot(np.arange(2, x), y, 'o-')
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.xticks(np.arange(2, x))
    plt.show()
```

```python
plot(max_range, scores, "No. of clusters", "Silhouette Score")
```

Now the data is K-Means clustered through the K-Means method and plotted on a *Folium* map:

```python
# set number of clusters according to graph
kclusters = 2

sf_grouped_clustering = sf_grouped.drop('Neighborhood', 1)

# run k-means clustering
kmeans = KMeans(n_clusters=kclusters, random_state=0).fit(sf_grouped_clustering)

# check cluster labels generated for each row in the dataframe
kmeans.labels_[0:10] 
```

```python
# add clustering labels
#neighborhoods_venues_sorted.drop(columns = {'Cluster Labels'})
neighborhoods_venues_sorted.insert(0, 'Cluster Labels', kmeans.labels_)

sf_merged = sf_df

# merge toronto_grouped with toronto_data to add latitude/longitude for each neighborhood
sf_merged = sf_merged.join(neighborhoods_venues_sorted.set_index('Neighborhood'), on='Neighborhood')

display(sf_merged.head())
```

```python
# The initializing the base of the map
address = 'San Francisco, California'

geolocator = Nominatim(user_agent="sf_view")
location = geolocator.geocode(address)
latitude = location.latitude
longitude = location.longitude
print('The geograpical coordinate of San Francisco are {}, {}.'.format(latitude, longitude))
```

```python
# Matplotlib and associated plotting modules
import matplotlib.cm as cm
import matplotlib.colors as colors

#!conda install -c conda-forge folium=0.5.0 --yes
import folium # map rendering library

# create map
map_clusters = folium.Map(location=[latitude, longitude], zoom_start=12)

# set color scheme for the clusters
x = np.arange(kclusters)
ys = [i + x + (i*x)**2 for i in range(kclusters)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# add markers to the map
markers_colors = []
for lat, lon, poi, cluster in zip(sf_merged['Latitude'], sf_merged['Longitude'], sf_merged['Neighborhood'], sf_merged['Cluster Labels']):
    label = folium.Popup(str(poi) + ' Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(map_clusters)
       
map_clusters
```
